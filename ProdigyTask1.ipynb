```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
titanic = pd.read_csv('tested.csv')
```

```python
# Calculate the distribution of ages
age_distribution = titanic['Age'].value_counts(normalize=True)

# Display the distribution of ages
print(age_distribution)
```

```python
# Create a histogram to visualize the distribution of ages
plt.hist(age_distribution.index, weights=age_distribution, bins=30)

# Add title and labels
plt.title('Distribution of Ages Among Passengers on the Titanic')
plt.xlabel('Age')
plt.ylabel('Percentage')

# Display the histogram
plt.show()
```
